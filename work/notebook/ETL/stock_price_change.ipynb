```python
import pandas as pd
```

```python
df_long = pd.read_csv("../../data/transform/non_esg_stock_prices_long_2022-01-17.csv")
display(df_long.ticker.nunique())
df_long
```

```python
df_wide = pd.read_csv("../../data/transform/non_esg_stock_prices_wide_2022-01-17.csv")
df_wide
```

price change % =  (today price - last biz date price) *100
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pct_change.html

```python
# Get full list of tickers 
ticker_list = df_wide.columns.tolist()[3:]
display(len(ticker_list))
ticker_list[0], ticker_list[-1]
```

```python
# Get the list of price % change 
aap_pc = []
for ticker in ticker_list:
    aap_pc=aap_pc+(df_wide[ticker].pct_change().tolist())
```

```python
display(len(df_long))
len(aap_pc)
```

```python
# Append to df 
df_long['pct_change'] = aap_pc 
```
